## Train Test Split



```python
X = emp_data.length_of_service
y = emp_data.age
```

```python
from sklearn.datasets import make_classification
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```


<h3>Random state for X, y for the classification will be 62,<br> while for train_test_split, random state will be 70 for getting more accuracy


```python
X, y = make_classification(random_state=62)
```


## Logistic Regression


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=70)
```

```python
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)  # apply scaling on training data
```

```python
lg_score = pipe.score(X_test, y_test)
print(f"\nAccuracy Score (Logistic Regression) between X_test and y_test : {lg_score}")
```

```python
log_model = LogisticRegression()
log_model = log_model.fit(X_train,y_train)
```

```python
 log_model.coef_
```

```python
 log_model.intercept_
```


## Decision Tree


```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=70)
```

```python
pipe = make_pipeline(StandardScaler(), DecisionTreeClassifier())
pipe.fit(X_train, y_train)  # apply scaling on training data
```

```python
dt_score = pipe.score(X_test, y_test)
print(f"\nAccuracy Score (Decision Tree) between X_test and y_test : {dt_score}")
```

```python
dt_model = LogisticRegression()
dt_model = dt_model.fit(X_train,y_train)
```

```python
 dt_model.coef_
```

```python
dt_model.intercept_
```

## Random Forest


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=70)
```

```python
pipe = make_pipeline(StandardScaler(), RandomForestClassifier())
pipe.fit(X_train, y_train)  # apply scaling on training data
```

```python
rd_score = pipe.score(X_test, y_test)
print(f"\nAccuracy Score (Random Forest) between X_test and y_test : {rd_score}")
```

```python
rdf_model = RandomForestClassifier()
rdf_model = rdf_model.fit(X_train,y_train)
```

```python
rdf_model.feature_importances_
```
